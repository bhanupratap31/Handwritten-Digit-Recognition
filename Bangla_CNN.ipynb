```
#Importing Required packages
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import Flatten
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from keras.utils import np_utils
from sklearn.metrics import classification_report, confusion_matrix
import os
from google.colab import drive
drive.mount("/content/gdrive", force_remount=True)

#Create the CNN Model
classifier = Sequential()
classifier.add(Conv2D(32,3,3,input_shape=(100,100,3),activation='relu'))
classifier.add(MaxPooling2D(pool_size=(2,2)))
classifier.add(Conv2D(32,3,3,activation='relu'))
classifier.add(MaxPooling2D(pool_size=(2,2)))
classifier.add(Flatten())
classifier.add(Dense(output_dim=128, activation='relu'))
classifier.add(Dense(output_dim=10, activation='softmax'))

#Import training images
train_dir0=os.path.join('/content/gdrive/My Drive/Bangla Dataset/New Data/Training/0')
train_dir1=os.path.join('/content/gdrive/My Drive/Bangla Dataset/New Data/Training/1')
train_dir2=os.path.join('/content/gdrive/My Drive/Bangla Dataset/New Data/Training/2')
train_dir3=os.path.join('/content/gdrive/My Drive/Bangla Dataset/New Data/Training/3')
train_dir4=os.path.join('/content/gdrive/My Drive/Bangla Dataset/New Data/Training/4')
train_dir5=os.path.join('/content/gdrive/My Drive/Bangla Dataset/New Data/Training/5')
train_dir6=os.path.join('/content/gdrive/My Drive/Bangla Dataset/New Data/Training/6')
train_dir7=os.path.join('/content/gdrive/My Drive/Bangla Dataset/New Data/Training/7')
train_dir8=os.path.join('/content/gdrive/My Drive/Bangla Dataset/New Data/Training/8')
train_dir9=os.path.join('/content/gdrive/My Drive/Bangla Dataset/New Data/Training/9')

#Import testing images
test_dir0=os.path.join('/content/gdrive/My Drive/Bangla Dataset/New Data/Testing/0')
test_dir1=os.path.join('/content/gdrive/My Drive/Bangla Dataset/New Data/Testing/1')
test_dir2=os.path.join('/content/gdrive/My Drive/Bangla Dataset/New Data/Testing/2')
test_dir3=os.path.join('/content/gdrive/My Drive/Bangla Dataset/New Data/Testing/3')
test_dir4=os.path.join('/content/gdrive/My Drive/Bangla Dataset/New Data/Testing/4')
test_dir5=os.path.join('/content/gdrive/My Drive/Bangla Dataset/New Data/Testing/5')
test_dir6=os.path.join('/content/gdrive/My Drive/Bangla Dataset/New Data/Testing/6')
test_dir7=os.path.join('/content/gdrive/My Drive/Bangla Dataset/New Data/Testing/7')
test_dir8=os.path.join('/content/gdrive/My Drive/Bangla Dataset/New Data/Testing/8')
test_dir9=os.path.join('/content/gdrive/My Drive/Bangla Dataset/New Data/Testing/9')

#Import the package for preprocessing the dataset
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1./255, shear_range=0.2, zoom_range=0.2, horizontal_flip=True) #Preprocess Train set
test_datagen = ImageDataGenerator(rescale=1./255) #Preprocess Test Set

#Now import the dataset from the directory
train_set=train_datagen.flow_from_directory('/content/gdrive/My Drive/Bangla Dataset/New Data/Training', target_size=(100,100), batch_size= 32, class_mode='categorical') #Problem is Here
test_set=train_datagen.flow_from_directory('/content/gdrive/My Drive/Bangla Dataset/New Data/Testing', target_size=(100,100), batch_size=32, class_mode='categorical') #Problem is Here

```

```
#Compile the Model
classifier.compile(optimizer= 'adam', loss='categorical_crossentropy', metrics=['accuracy'])

#Fit the dataset to the classifier and run the model
history=classifier.fit_generator(train_set, samples_per_epoch=593, nb_epoch=20, validation_data=test_set, nb_val_samples=200)
```

```
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report 
import numpy as np
Y_pred = classifier.predict_generator(test_set, 200 // 32 +1) #no.of test samples//batch_size+1
y_pred = np.argmax(Y_pred, axis=1)
print('Confusion Matrix')
print('----------------')
print(confusion_matrix(test_set.classes, y_pred))
print('Classification report')
print('----------------')
print(classification_report(test_set.classes, y_pred))
```

```
import matplotlib.pyplot as plt

#Plot the Loss Curves
plt.figure(figsize=[8,6])
plt.plot(history.history['loss'],'g',linewidth=3.0)
plt.plot(history.history['val_loss'],'b',linewidth=3.0)
plt.legend(['Training loss', 'Validation Loss'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Loss',fontsize=16)
plt.title('Loss Curves',fontsize=16)

#Plot the Accuracy Curves
plt.figure(figsize=[8,6])
plt.plot(history.history['acc'],'g',linewidth=3.0)
plt.plot(history.history['val_acc'],'b',linewidth=3.0)
plt.legend(['Training Accuracy', 'Validation Accuracy'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Accuracy',fontsize=16)
plt.title('Accuracy Curves',fontsize=16)
```
